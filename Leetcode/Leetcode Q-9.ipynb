```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        Max_Limit = 2**31 - 1
        Min_Limit = -2**31
        if x<0 or x>=Max_Limit or x<=Min_Limit:
            return False
        if x>=0:
            rev = 0
            num = x
            while x > 0:
                digit = x%10
                rev = rev*10+digit
                x=x//10
            if(rev == num):
                return True
            else:
                return False
```

```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        Max_Limit = 2**31 - 1
        Min_Limit = -2**31
        if x<0 or x>=Max_Limit or x<=Min_Limit:
            return False
        x = abs(x)
        rev = int(str(x)[::-1])
        return rev == x
```
