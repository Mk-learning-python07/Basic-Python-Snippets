### Leetcode Q-169. Majority Element

Difficulty - Easy<br>
Given an array nums of size n, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.



##### Problem Statement in detail : [link](https://leetcode.com/problems/majority-element/description/)
##### Answer Logic:
Find the candidate by looping through the array once and count the number of times you are getting the number. Then verify the found candidate by checking the criteria provided.


```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        maj_element = None
        count = 0

        for num in nums:
            if count == 0:
                maj_element = num
                count = 1
            elif num == maj_element:
                count += 1
            else:
                count -= 1
    
        # Second pass: Verify the candidate
        count = 0
        for num in nums:
            if num == maj_element:
                count += 1
        
        if count > len(nums) // 2:
            return maj_element
        else:
            raise ValueError("No majority element found")
```
