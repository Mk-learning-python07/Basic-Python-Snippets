## Leet Code Question - 1. Two Sum
Difficulty - Easy

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.


##### Problem Statement in detail : [link](https://leetcode.com/problems/two-sum/description/)
##### Answer Logic:

We try to use enumerate to figure out the combination of two numbers that is equal to target by iterating via number1.
#### number2 = target - number1

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        add = {}
        for i,j in enumerate(nums):
            diff = target - nums[i]
            if diff in add:
                    return [i,add[diff]]
            else:
                 add[j] = i
```
