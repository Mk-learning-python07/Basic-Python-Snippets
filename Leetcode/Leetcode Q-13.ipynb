### Leetcode Q-13. Roman to Integer

Difficulty - Medium

Seven different symbols represent Roman numerals with the following values:




For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

    I can be placed before V (5) and X (10) to make 4 and 9. 
    X can be placed before L (50) and C (100) to make 40 and 90. 
    C can be placed before D (500) and M (1000) to make 400 and 900.

Given a roman numeral, convert it to an integer.



##### Problem Statement in detail : [link](https://leetcode.com/problems/roman-to-integer/)
##### Answer Logic:

Follow Q-12 logic in reverse - This will convert Roman Numeral to Integer. Since we have IV IX etc where the number previous is less than the current number, we need to subtract that from the previous figure. Once can achieve that by subtracting the previous number 2 times from the current number that is getting added to the list.

Ex: IX = 1+10-2*1 = 9.


```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        roman_to_int_map ={
        'I': 1, 'V': 5, 'X': 10, 'L': 50,
        'C': 100, 'D': 500, 'M': 1000
        }
        
        n= 0
        n_1 = 0
    
        for char in s:
            j = roman_to_int_map[char]
            if j > n_1:
                n += j - 2 * n_1
            else:
                n += j
            n_1 = j
        return n
```
