### Leetcode Q-2011. Final Value of Variable After Performing Operations

Difficulty - Easy<br>
There is a programming language with only four operations and one variable X:

    ++X and X++ increments the value of the variable X by 1.
    --X and X-- decrements the value of the variable X by 1.

Initially, the value of X is 0.

Given an array of strings operations containing a list of operations, return the final value of X after performing all the operations.


```python
class Solution(object):
    def finalValueAfterOperations(self, operations):
        """
        :type operations: List[str]
        :rtype: int
        """
        x =operations.count("++X")
        x +=operations.count("X++")
        x -=operations.count("--X") 
        x -=operations.count("X--")
        return x
```

### Leetcode Q-1512. Number of Good Pairs

Difficulty - Easy<br>

Given an array of integers nums, return the number of good pairs.

A pair (i, j) is called good if nums[i] == nums[j] and i < j.



```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        Op = 0
        for i in range(len(nums)):
            for j in range(len(nums)):
                if (nums[i] == nums[j] and i<j):
                    Op += 1

        return Op
```

### Leetcode Q-1470. Shuffle the Array
Difficulty - Easy<br>

Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

Return the array in the form [x1,y1,x2,y2,...,xn,yn].


```python
class Solution(object):
    def shuffle(self, nums, n):
        """
        :type nums: List[int]
        :type n: int
        :rtype: List[int]
        """
        new = []
        i = 0
        for i in range(len(nums)//2):
            new.append(nums[i])
            new.append(nums[n+i])
        return new
```

### Leetcode Q-2942. Find Words Containing Character

Difficulty - Easy<br>

You are given a 0-indexed array of strings words and a character x.

Return an array of indices representing the words that contain the character x.

Note that the returned array may be in any order.

Example 1:

Input: words = ["leet","code"], x = "e"
Output: [0,1]
Explanation: "e" occurs in both words: "leet", and "code". Hence, we return indices 0 and 1.



```python
class Solution(object):
    def findWordsContaining(self, words, x):
        """
        :type words: List[str]
        :type x: str
        :rtype: List[int]
        """
        a = []
        for i,word in enumerate(words):
            if x in word:
                a.append(i)  
        return a
```

### Leetcode Q-1672. Richest Customer Wealth

Difficulty - Easy<br>

You are given an m x n integer grid accounts where accounts[i][j] is the amount of money the ith customer has in the jth bank. Return the wealth that the richest customer has.

A customer's wealth is the amount of money they have in all their bank accounts. The richest customer is the customer that has the maximum wealth.



```python
class Solution(object):
    def maximumWealth(self, accounts):
        """
        :type accounts: List[List[int]]
        :rtype: int
        """
        row_sums = []
        for row in accounts:
            sum_row = sum(row)
            row_sums.append(sum_row)
        row_sums.sort(reverse = True)
        return row_sums[0]
```

### Leetcode Q-2373. Largest Local Values in a Matrix


Difficulty - Easy<br>

You are given an n x n integer matrix grid.

Generate an integer matrix maxLocal of size (n - 2) x (n - 2) such that:

    maxLocal[i][j] is equal to the largest value of the 3 x 3 matrix in grid centered around row i + 1 and column j + 1.

In other words, we want to find the largest value in every contiguous 3 x 3 matrix in grid.

Return the generated matrix. <br>


```python
class Solution(object):
    def largestLocal(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: List[List[int]]
        """
        n = len(grid)
        maxLocal = [[0] * (n - 2) for _ in range(n - 2)]
    
        for i in range(n - 2):
            for j in range(n - 2):
            # Calculate the maximum value in the 3x3 submatrix centered at (i+1, j+1)
                max_val = float('-inf')
                for r in range(3):
                    for c in range(3):
                        max_val = max(max_val, grid[i + r][j + c])
                
                maxLocal[i][j] = max_val
        return maxLocal
```

### Leetcode Q-1863. Sum of All Subset XOR Totals

Difficulty - Easy<br>

The XOR total of an array is defined as the bitwise XOR of all its elements, or 0 if the array is empty.

    For example, the XOR total of the array [2,5,6] is 2 XOR 5 XOR 6 = 1.

Given an array nums, return the sum of all XOR totals for every subset of nums. 

Note: Subsets with the same elements should be counted multiple times.

An array a is a subset of an array b if a can be obtained from b by deleting some (possibly zero) elements of b.



```python
class Solution(object):
    def subsetXORSum(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        def backtrack(index, current_xor):
            if index == len(nums):
                return current_xor
            
            # Include nums[index] in the subset
            include_xor = backtrack(index + 1, current_xor ^ nums[index])
            
            # Exclude nums[index] from the subset
            exclude_xor = backtrack(index + 1, current_xor)
            
            return include_xor + exclude_xor
        
        return backtrack(0, 0)        
```

### Leetcode Q-2037. Minimum Number of Moves to Seat Everyone

Difficulty - Easy<br>
There are n availabe seats and n students standing in a room. You are given an array seats of length n, where seats[i] is the position of the ith seat. You are also given the array students of length n, where students[j] is the position of the jth student.

You may perform the following move any number of times:

    Increase or decrease the position of the ith student by 1 (i.e., moving the ith student from position x to x + 1 or x - 1)

Return the minimum number of moves required to move each student to a seat such that no two students are in the same seat.

Note that there may be multiple seats or students in the same position at the beginning.




```python
class Solution(object):
    def minMovesToSeat(self, seats, students):
        """
        :type seats: List[int]
        :type students: List[int]
        :rtype: int
        """
        seats.sort()
        students.sort()
        
        n = len(seats)
        moves = 0
        
        for i in range(n):
            moves += abs(students[i] - seats[i])
        
        return moves
```

### Leetcode Q-2798. Number of Employees Who Met the Target

Difficulty - Easy<br>

There are n employees in a company, numbered from 0 to n - 1. Each employee i has worked for hours[i] hours in the company.

The company requires each employee to work for at least target hours.

You are given a 0-indexed array of non-negative integers hours of length n and a non-negative integer target.

Return the integer denoting the number of employees who worked at least target hours.


```python
class Solution(object):
    def numberOfEmployeesWhoMetTarget(self, hours, target):
        """
        :type hours: List[int]
        :type target: int
        :rtype: int
        """
        n = 0
        for i in hours:
            if i>=target:
                n += 1
        return n
```

### Leetcode Q-1431. Kids With the Greatest Number of Candies


Difficulty - Easy<br>
There are n kids with candies. You are given an integer array candies, where each candies[i] represents the number of candies the ith kid has, and an integer extraCandies, denoting the number of extra candies that you have.

Return a boolean array result of length n, where result[i] is true if, after giving the ith kid all the extraCandies, they will have the greatest number of candies among all the kids, or false otherwise.

Note that multiple kids can have the greatest number of candies.



```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        """
        :type candies: List[int]
        :type extraCandies: int
        :rtype: List[bool]
        """
        R = []
        max_c = max(candies)
        for i in range(len(candies)):
            R.append(candies[i]+extraCandies>=max_c)
        return R
```

### Leetcode Q-1637. Widest Vertical Area Between Two Points Containing No Points

Difficulty - Easy<br>

Given n points on a 2D plane where points[i] = [xi, yi], Return the widest vertical area between two points such that no points are inside the area.

A vertical area is an area of fixed-width extending infinitely along the y-axis (i.e., infinite height). The widest vertical area is the one with the maximum width.

Note that points on the edge of a vertical area are not considered included in the area.



```python
class Solution(object):
    def maxWidthOfVerticalArea(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """
        # Extract x-coordinates
        x_coordinates = [point[0] for point in points]
        
        # Sort x-coordinates
        x_coordinates.sort()
        
        # Initialize maximum width to 0
        max_width = 0
        
        # Calculate maximum gap between consecutive x-coordinates
        for i in range(1, len(x_coordinates)):
            max_width = max(max_width, x_coordinates[i] - x_coordinates[i - 1])
        
        return max_width
```
