### Leetcode Q-8. String to Integer (atoi)


Difficulty - Medium

Implement the myAtoi(string s) function, which converts a string to a 32-bit signed integer.

The algorithm for myAtoi(string s) is as follows:

    Whitespace: Ignore any leading whitespace (" ").
    Signedness: Determine the sign by checking if the next character is '-' or '+', assuming positivity is neither present.
    Conversion: Read the integer by skipping leading zeros until a non-digit character is encountered or the end of the string is reached. If no digits were read, then the result is 0.
    Rounding: If the integer is out of the 32-bit signed integer range [-231, 231 - 1], then round the integer to remain in the range. Specifically, integers less than -231 should be rounded to -231, and integers greater than 231 - 1 should be rounded to 231 - 1.

Return the integer as the final result.



##### Problem Statement in detail : [link](https://leetcode.com/problems/string-to-integer-atoi/description/)
##### Answer Logic:

I found the answer in 4 steps as the problem dictated.
### Step - 1: Strip any extra whitespaces
### Step - 2: Read the first character to interpret - or +. Remove the first part from the string
### Step - 3: Find the integer part. This is where the characters are checked for integers and loop breaks if a word is encountered.
### Step - 4: Return the final integer value

```python
class Solution(object):
    def myAtoi(self, s):
        """
        :type s: str
        :rtype: int
        """
        # Step - 1: Strip any extra whitespaces
        s = s.strip(" ")
        flag = 1
        if not s:
            return 0
        # Step - 2: Read the first character to interpret - or + . Remove the character from the string.
        if s[0] == '-':
            flag = -1  # Negative sign encountered, set flag to -1
            s = s[1:]  
        elif s[0] == '+':
            s = s[1:]  
        
        # Step - 3: Find the integer part
        n = 0
        for i in s:
            if i.isdigit():
                n = n * 10 + int(i)
                # Check for overflow based on the problem constraints
                if flag * n > 2**31 - 1:
                    return 2**31 - 1
                elif flag * n < -2**31:
                    return -2**31
            else:
                break  # Stop further processing if non-digit character is encountered
        
        # Step - 4: Return the final integer value
        return n * flag
```
