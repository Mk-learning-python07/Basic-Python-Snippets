### Leetcode Q-7. Reverse Integer

Difficulty - Medium

Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-2^31, 2^31 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).

##### Answer Logic:

I was able to find two solutions. 
The first one was math based. A number is reversed when we start taking the digits in reverse and start multiplying it by powers of 10 in the descending order. So we try to achieve it in a looped manner by isolating the last digit first and consecutively adding it to the reverse sum. We find the number to be negative by using a flag variable and multiply it to the absolute number which is reversed. The overflow conditions come at last as the problem stipulated.

Second solution was kind of ingenious by using a string array and using splicing to reverse without any loops. Credits to codewithmohitofficial - Leetcode user for explaining this idea.


```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        rev = 0
        flag = 1
        if x < 0:
            flag = -1

        x = abs(x)
        while int (x) > 0:
            digit = x%10
            #print(f'digit:',digit)
            rev = rev*10+digit
            x=x//10
            #print(f'rev:',rev)
        if not (rev <= pow(2,31)-1 and rev >= pow(-2,31)):
            rev = 0
        return rev*flag
```

```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        flag = 1
        if (x<0):
            flag = -1
        x = abs(x)
        rev = int(str(x)[::-1])
        if not (rev <= pow(2,31)-1 and rev >= pow(-2,31)):
            rev = 0
        return rev*flag
```
