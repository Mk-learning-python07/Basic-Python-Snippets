## Leet Code Question - 5. Longest Palindromic Substring

Difficulty - Medium

Given a string s, return the longest palindromic substring in s.


Example 1:

Input: s = "babad"
Output: "bab"
Explanation: "aba" is also a valid answer.


##### Problem Statement in detail : [link](https://leetcode.com/problems/longest-palindromic-substring/description/)
##### Answer Logic:

A palindrome is a word that looks the same if spelt backwards as shown in the image. So to figure out a palindromic substring, you search a string from its center and move equally on left and right from the center string to check if they are the same - This will ensure they are same if spelt backwards.<br>



```python
class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        def search_pal(left, right):
            while left >= 0 and right < len(s) and s[left] == s[right]:
                
                left -= 1
                right += 1
            return s[left+1:right]

        sub = ""
        for i in range(len(s)):
            # Odd length palindromes (single character center)
            sub1 = search_pal(i, i)
            if len(sub1)>len(sub):
                sub = sub1
            
            # Even length palindromes (two consecutive characters center)
            sub1 = search_pal(i, i + 1)
            if len(sub1) > len(sub):
                sub = sub1
        return sub
```
