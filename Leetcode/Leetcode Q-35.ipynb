### Leetcode Q-35. Search Insert Position


Difficulty - Easy<br>

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.
 
Example 1:

Input: nums = [1,3,5,6], target = 5
Output: 2

Example 2:

Input: nums = [1,3,5,6], target = 2
Output: 1

Example 3:

Input: nums = [1,3,5,6], target = 7
Output: 4



##### Problem Statement in detail : [link](https://leetcode.com/problems/search-insert-position/description/)
##### Answer Logic:

Search the array and find the location if the number is available. If not available, since the array is sorted find the number >=target and return its location.



```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        for i in range(len(nums)):
            if nums[i] == target:
                return i
            elif nums[i] >=target:
                return i
        return len(nums)
```
