### Leetcode Q-20. Valid Parentheses

Difficulty - Easy

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

    Open brackets must be closed by the same type of brackets.
    Open brackets must be closed in the correct order.
    Every close bracket has a corresponding open bracket of the same type.

 

Example 1:

Input: s = "()"
Output: true

Example 2:

Input: s = "()[]{}"
Output: true

Example 3:

Input: s = "(]"
Output: false



##### Problem Statement in detail : [link](https://leetcode.com/problems/valid-parentheses/description/)
##### Partial Answer Logic:
 If we count the number of braces = Adding 1 for open and subtracting 1 for close for different types of braces. Finally if we check all the three numbers, then we will know if there are equal number of open/closed braces. This failed because there is a closing order that needs to be followed.

Final Solution: Credits to veeannzhang - Leetcode user. I used their idea of stacking this solution and made the code a bit more efficient.
#### 1. if it's the left bracket then we append it to the stack
#### 2. else if it's the right bracket and the stack is empty(meaning no matching left bracket), or the left bracket doesn't match
#### 3. finally check if the stack length is zero.

```python
# Partial Solution
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        n = 0
        c = 0
        p = 0
        if(s==""):
            return False
        for i in range(len(s)):
            if(s[i] =='('):
                n += 1
            elif(s[i] == ')'):
                n -= 1
            elif(s[i] == '['):
                p +=1
            elif(s[i] == ']'):
                p -= 1
            elif(s[i] == '}'):
                c -= 1
            else:
                c += 1
        if(n==0 and p==0 and c==0):
            return True
        else:
            return False
```

```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        stack = []
        bracket_map = {')': '(', ']': '[', '}': '{'}
        
        for char in s:
            if char in '([{':
                stack.append(char)
            elif char in ')]}':
                if not stack or stack.pop() != bracket_map[char]:
                    return False
        
        return len(stack) == 0
```
