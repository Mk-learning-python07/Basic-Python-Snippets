### Leetcode Q-279. Perfect Squares

Difficulty - Medium<br>

Given an integer n, return the least number of perfect square numbers that sum to n.

A perfect square is an integer that is the square of an integer; in other words, it is the product of some integer with itself. For example, 1, 4, 9, and 16 are perfect squares while 3 and 11 are not.

 

Example 1:

Input: n = 12
Output: 3
Explanation: 12 = 4 + 4 + 4.

Example 2:

Input: n = 13
Output: 2
Explanation: 13 = 4 + 9.



##### Problem Statement in detail : [link](https://leetcode.com/problems/perfect-squares/description/)
##### Answer Logic:
We initialize a list where it will eventually store the minimum number of perfect squares summing up to i. dp[0] = 0 for handling n=0. The loop for i in range(1, n + 1): iterates through each number from 1 to n. Adding 1 accounts for the j*j perfect square added to the sum. Essentially, dp[i] is built using previously computed results for smaller numbers, leveraging the subproblem solutions stored in dp.The final array element ontains the minimum number of perfect squares that sum up to n, which is returned as the result.


```python
class Solution(object):
    def numSquares(self, n):
        """
        :type n: int
        :rtype: int
        """
        dp = [float('inf')] * (n + 1)
        dp[0] = 0  # 0 perfect squares sum to 0
    
         # Iterate through all numbers from 1 to n
        for i in range(1, n + 1):
        # Check all perfect squares j*j that are <= i
            j = 1
            while j * j <= i:
                dp[i] = min(dp[i], dp[i - j * j] + 1)
                j += 1
    
        return dp[n]
```
