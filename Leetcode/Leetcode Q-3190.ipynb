### Leetcode Q-3190. Find Minimum Operations to Make All Elements Divisible by Three
Easy


Difficulty - Easy<br>
You are given an integer array nums. In one operation, you can add or subtract 1 from any element of nums.

Return the minimum number of operations to make all elements of nums divisible by 3.

 

Example 1:

Input: nums = [1,2,3,4]

Output: 3



##### Problem Statement in detail : [link](https://leetcode.com/problems/find-minimum-operations-to-make-all-elements-divisible-by-three/description/)
##### Answer Logic:

We need to find numbers that are not divisibile by 3. The minimum operations for num%3!=0 (For remainder of 1 you need to subtract 1 and for remainder of 2 you need
to add 1)

```python
class Solution(object):
    def minimumOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return sum(num%3>0 for num in nums)
```
