### Leetcode Q-2108. Find First Palindromic String in the Array


Difficulty - Easy<br>
Given an array of strings words, return the first palindromic string in the array. If there is no such string, return an empty string "".

A string is palindromic if it reads the same forward and backward.




##### Problem Statement in detail : [link](https://leetcode.com/problems/find-first-palindromic-string-in-the-array/description/)
##### Answer Logic:
Write a subfunction to find if a string is palindrome or not and in the driver function pass individual words to figure out if it is a palindrome or not and break the loop once you find the first occurence.

```python
class Solution(object):
    def check_palindrome(self, s):
        left = 0
        right = len(s) - 1
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        
        return True

    def firstPalindrome(self, words):
        """
        :type words: List[str]
        :rtype: str
        """
        n = len(words)
        for i in range(n):
            if self.check_palindrome(words[i]):
                return words[i]
        return ""    
```
